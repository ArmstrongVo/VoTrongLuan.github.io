```python
from matplotlib import figure
#from skfuzzy.membership.generatemf import gaussmf
import matplotlib.pyplot as plt
#from skfuzzy import control as ctrl 
import numpy as np
#import skfuzzy as fuzz
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axis3d
import pandas as pd
url = 'insurance.csv'
data = pd.read_csv(url)
data.hist()
plt.show 
```

```python
correlation = data.corr() # mối tương quan giữa các biến, trị số lớn nhất là ảnh hưởng nhiều nhất, TLT HAY TLN  phụ thuộc vào dấu + or -
print(correlation)
import seaborn as sns   # thư viện hiển thị dữ liệu
#sns.pairplot(data)
sns.heatmap(correlation,annot=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
X=pd.DataFrame([data['age'],data['age']**2,data['age']**3]).T
model.fit(X,y=data['charges'])
print('intercept của mô hình',model.intercept_)
print('coef của mô hình',model.coef_)
print('độ chính xác của mô hình',model.score(X,y=data['charges']))
data.plot(kind = 'scatter', x = 'age', y='charges',figsize = (9,9), color = 'black',xlim=(0,60))
range = np.arange(1.5,50.5,0.1)
predictor = pd.DataFrame([range,range**2,range**3]).T
y_pred=model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
